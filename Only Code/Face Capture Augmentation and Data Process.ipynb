# Taking Frontal Face picture using camera and saving it into the dataset
# Code for Capturing Frontal,Right,Left-Side Faces:

```python
import cv2
import os
import dlib
import numpy as np
import time
import shutil

# Initialize dlib's face detector (HOG-based) and shape predictor
detector = dlib.get_frontal_face_detector()
shape_predictor_path = r"F:\All Face Recognition Thesis\Simple CNN with Dynamic Fully Connected Layer\shape_predictor_68_face_landmarks.dat"
predictor = dlib.shape_predictor(shape_predictor_path)

def process_image(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return gray_image

def face_alignment(image, landmarks):
    # Placeholder for face alignment logic
    return image

def draw_landmarks(image, landmarks):
    for (x, y) in landmarks:
        cv2.circle(image, (x, y), 2, (0, 255, 0), -1)
    return image

def capture_faces(face_position, person_name, person_id):
    print(f"\nPlease position your face for {face_position} face capture.")
    input("Press Enter when ready...")

    # Initialize the camera
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("No camera found. Please connect a camera and try again.")
        return False

    # Create the base directory for storing images and landmarks
    base_dir = f"Images/{person_name}_{person_id}"
    os.makedirs(base_dir, exist_ok=True)

    img_id = len(os.listdir(base_dir)) + 1
    captured_images = 0  # Counter for captured images

    while captured_images < 20:  # Limit to 20 images
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image. Exiting...")
            break

        gray = process_image(frame)
        faces = detector(gray)

        if face_position == "Frontal":
            if len(faces) == 0:
                print("No faces detected. Please adjust your position.")
                continue

            for face in faces:
                landmarks = predictor(gray, face)
                landmarks_np = np.array([[p.x, p.y] for p in landmarks.parts()])
                frame_with_landmarks = draw_landmarks(frame.copy(), landmarks_np)
                aligned_face = face_alignment(frame, landmarks_np)

                if aligned_face is not None:
                    img_id += 1
                    aligned_face_gray = cv2.cvtColor(aligned_face, cv2.COLOR_BGR2GRAY)
                    face_resized = cv2.resize(aligned_face_gray, (200, 200))

                    file_name_path = f"{base_dir}/{person_name}_{person_id}.{img_id}_{face_position}.jpg"
                    cv2.imwrite(file_name_path, face_resized)

                    cv2.putText(face_resized, f"{img_id}", (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
                    cv2.imshow(f"Aligned Face ({face_position})", face_resized)

                    cv2.putText(frame_with_landmarks, f"Face {face_position} Capture: OK", (10, 30), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
                    cv2.imshow(f"Face with Landmarks ({face_position})", frame_with_landmarks)

                    captured_images += 1  # Increment counter for captured images

        else:
            # For left and right face captures, directly save the images without face detection
            img_id += 1
            aligned_face_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            face_resized = cv2.resize(aligned_face_gray, (200, 200))

            file_name_path = f"{base_dir}/{person_name}_{person_id}.{img_id}_{face_position}.jpg"
            cv2.imwrite(file_name_path, face_resized)

            cv2.putText(face_resized, f"{img_id}", (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
            cv2.imshow(f"Aligned Face ({face_position})", face_resized)

            captured_images += 1  # Increment counter for captured images

        if cv2.waitKey(1) == 13:  # Break on 'Enter' key press
            break

    # Release the camera and close OpenCV windows
    cap.release()
    cv2.destroyAllWindows()

    if captured_images > 0:
        print(f"{captured_images} images captured for {face_position} face.")
        return True
    else:
        return False

def generate_dataset():
    # Prompt user for the person's name and ID
    person_name = input("Enter the person's name: ").strip()
    person_id = input("Enter the person's ID: ").strip()

    # Capture frontal face
    if not capture_faces("Frontal", person_name, person_id):
        print("Frontal face capture failed. Exiting...")
        return

    # Capture left side face
    if not capture_faces("Left_Side", person_name, person_id):
        print("Left side face capture failed. Exiting...")
        return

    # Capture right side face
    if not capture_faces("Right_Side", person_name, person_id):
        print("Right side face capture failed. Exiting...")
        return

    print(f"Dataset creation complete for {person_name} with ID {person_id}.")

generate_dataset()
```

```python
import cv2
import os
import dlib
import numpy as np
import shutil  # For folder deletion
from pathlib import Path

# Initialize dlib's face detector (HOG-based) and shape predictor
detector = dlib.get_frontal_face_detector()
shape_predictor_path = r"F:\All Face Recognition Thesis\Simple CNN with Dynamic Fully Connected Layer\shape_predictor_68_face_landmarks.dat"

# Check if the shape predictor file exists
if not os.path.exists(shape_predictor_path):
    raise FileNotFoundError(f"Shape predictor file not found at {shape_predictor_path}")

predictor = dlib.shape_predictor(shape_predictor_path)

def process_image(image):
    """Convert the input image to grayscale."""
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def face_alignment(image, landmarks):
    """Placeholder for face alignment logic using landmarks."""
    return image  # Add your actual alignment code here if necessary

def process_images_in_folder(person_folder_path, person_folder_name):
    """Process all images in the given person's folder."""
    img_id = 0
    faces_detected = False  # Flag to track if any faces are detected
    base_dir = f"Images/{person_folder_name}"  # Directory to store processed images
    os.makedirs(base_dir, exist_ok=True)

    for image_name in os.listdir(person_folder_path):
        if image_name.lower().endswith((".jpg", ".png", ".jpeg")):  # Added support for .jpeg
            image_path = os.path.join(person_folder_path, image_name)
            print(f"Processing image: {image_path}")

            # Read the image
            image = cv2.imread(image_path)
            if image is None:
                print(f"Failed to read image: {image_name}. Skipping...")
                continue

            # Convert image to grayscale
            gray = process_image(image)

            # Detect faces
            faces = detector(gray)
            print(f"Faces detected in {image_name}: {len(faces)}")

            if len(faces) == 0:
                print(f"No faces detected in {image_name}.")
                continue

            faces_detected = True  # Set flag to True if faces are detected

            for face in faces:
                # Predict facial landmarks
                try:
                    landmarks = predictor(gray, face)
                except Exception as e:
                    print(f"Error predicting landmarks in {image_name}: {e}")
                    continue

                landmarks_np = np.array([[p.x, p.y] for p in landmarks.parts()])

                # Perform face alignment (if applicable)
                aligned_face = face_alignment(image, landmarks_np)

                if aligned_face is not None:
                    img_id += 1

                    # Convert the aligned face to grayscale
                    aligned_face_gray = cv2.cvtColor(aligned_face, cv2.COLOR_BGR2GRAY)

                    # Resize face to 200x200 pixels
                    face_resized = cv2.resize(aligned_face_gray, (200, 200))

                    # Save processed face image
                    file_name_path = f"{base_dir}/{person_folder_name}.{img_id}.jpg"
                    cv2.imwrite(file_name_path, face_resized)

                    # Display the processed face with ID (optional)
                    cv2.putText(face_resized, f"{img_id}", (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
                    cv2.imshow("Processed Face", face_resized)
                    cv2.waitKey(100)  # Display the image for 100ms (optional)

    # Close the display window
    cv2.destroyAllWindows()

    return faces_detected  # Return whether faces were detected or not

def process_existing_images():
    """Main function to process all folders in the image upload directory."""
    image_upload_dir = "imageUpload"
    
    if not os.path.exists(image_upload_dir):
        raise FileNotFoundError(f"Image upload directory not found: {image_upload_dir}")
    
    # Iterate over all folders in the image upload directory
    for person_folder in os.listdir(image_upload_dir):
        person_folder_path = os.path.join(image_upload_dir, person_folder)

        # Skip if not a directory
        if not os.path.isdir(person_folder_path):
            continue

        print(f"Processing folder: {person_folder_path}")

        # Check if the folder already has a tracking file (processed.txt)
        tracking_file = os.path.join(person_folder_path, "processed.txt")
        if os.path.exists(tracking_file):
            print(f"Folder '{person_folder}' has already been processed. Skipping...")
            continue

        # Process images in the current person's folder
        faces_detected = process_images_in_folder(person_folder_path, person_folder)

        if faces_detected:
            print(f"Processing complete for {person_folder}!")

            # Create a tracking file to mark as processed
            with open(tracking_file, "w") as file:
                file.write("Processed")
        else:
            print(f"No faces were detected in {person_folder}. Processing aborted.")
            # Delete the folder if no faces were detected
            base_dir = f"Images/{person_folder}"
            if os.path.exists(base_dir):
                shutil.rmtree(base_dir)
                print(f"Deleted folder: {base_dir}")

process_existing_images()
```

# DATA AUGMENTATION
# Turning the entire Datatset into grayscale and cropping only faces(pytorch done) [68 landmarks]Â¶

```python
import cv2
import os
import numpy as np
import random
from PIL import Image
import dlib

# Initialize dlib's face detector and shape predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(r"F:\All Face Recognition Thesis\Simple CNN with Dynamic Fully Connected Layer\shape_predictor_68_face_landmarks.dat")

# Function to save landmarks as .npy files
def save_landmarks(landmarks, output_dir, img_id):
    landmarks_file = os.path.join(output_dir, f"{img_id}_landmarks.npy")
    np.save(landmarks_file, landmarks)
    print(f"Landmarks saved: {landmarks_file}")

# Function to resize image while maintaining aspect ratio
def resize_with_aspect_ratio(image, new_width):
    h, w = image.shape[:2]
    aspect_ratio = h / w
    new_height = int(new_width * aspect_ratio)
    return cv2.resize(image, (new_width, new_height))

# Function for random horizontal flip
def random_horizontal_flip(image, prob=0.5):
    if random.random() < prob:
        return cv2.flip(image, 1)  # Flip image horizontally
    return image

# Function for random rotation
def random_rotation(image, angle_range=(-15, 15)):
    angle = random.uniform(angle_range[0], angle_range[1])
    h, w = image.shape[:2]
    center = (w // 2, h // 2)
    rot_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    return cv2.warpAffine(image, rot_matrix, (w, h))

# Function for random zoom (scaling)
def random_zoom(image, zoom_range=(0.8, 1.2)):
    zoom_factor = random.uniform(zoom_range[0], zoom_range[1])
    h, w = image.shape[:2]
    new_h, new_w = int(h * zoom_factor), int(w * zoom_factor)
    resized_image = cv2.resize(image, (new_w, new_h))

    # Crop back to original size
    if zoom_factor < 1:  # Padding if zoom out
        pad_h = (h - new_h) // 2
        pad_w = (w - new_w) // 2
        return cv2.copyMakeBorder(resized_image, pad_h, pad_h, pad_w, pad_w, cv2.BORDER_CONSTANT, value=[0, 0, 0])
    else:  # Center crop if zoom in
        start_h = (new_h - h) // 2
        start_w = (new_w - w) // 2
        return resized_image[start_h:start_h + h, start_w:start_w + w]

# Function for random brightness and contrast adjustment
def random_brightness_contrast(image, brightness_range=(0.7, 1.3), contrast_range=(0.7, 1.3)):
    brightness = random.uniform(brightness_range[0], brightness_range[1])
    contrast = random.uniform(contrast_range[0], contrast_range[1])
    image = cv2.convertScaleAbs(image, alpha=contrast, beta=brightness * 50)  # Adjust contrast and brightness
    return image

# Use frontal face detector for frontal, and profile face detector for left/right profile faces
profile_face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_profileface.xml')

def process_folder(base_dir, output_dir, new_width):
    for folder_name in os.listdir(base_dir):
        folder_path = os.path.join(base_dir, folder_name)
        output_person_dir = os.path.join(output_dir, folder_name)

        if os.path.isdir(folder_path):
            print(f"Processing folder: {folder_name}")
            os.makedirs(output_person_dir, exist_ok=True)

            track_file_path = os.path.join(output_person_dir, f"{folder_name}_processed.txt")
            if os.path.exists(track_file_path):
                print(f"Skipping folder {folder_name} as it has already been processed.")
                continue

            img_id = 1
            for image_name in os.listdir(folder_path):
                image_path = os.path.join(folder_path, image_name)
                if image_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image = cv2.imread(image_path)
                    if image is None:
                        print(f"Failed to load image: {image_path}")
                        continue

                    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                    gray = resize_with_aspect_ratio(gray, new_width)

                    # Try frontal face detection first
                    faces = detector(gray)
                    
                    # If no faces detected, try profile face detection
                    if len(faces) == 0:
                        profile_faces = profile_face_cascade.detectMultiScale(gray, 1.1, 5)
                        for (x, y, w, h) in profile_faces:
                            faces.append(dlib.rectangle(x, y, x+w, y+h))  # Convert to dlib rectangle

                    # If still no faces detected, treat it as a left/right profile image
                    if len(faces) == 0:
                        print(f"No faces detected in {image_path}. Treating as side profile.")
                        faces = [dlib.rectangle(0, 0, gray.shape[1], gray.shape[0])]  # Assume whole image is face

                    for face in faces:
                        landmarks = predictor(gray, face)
                        landmarks_np = np.array([[p.x, p.y] for p in landmarks.parts()])
                        save_landmarks(landmarks_np, output_person_dir, img_id)

                        # Crop only the face region
                        face_crop = gray[face.top():face.bottom(), face.left():face.right()]

                        # Ensure the face crop is valid
                        if face_crop.size > 0:
                            face_resized = cv2.resize(face_crop, (200, 200))

                            # Apply augmentations only to the face region
                            augmented_images = [
                                random_horizontal_flip(face_resized),
                                random_rotation(face_resized),
                                random_zoom(face_resized),
                                random_brightness_contrast(face_resized)
                            ]

                            for idx, aug_image in enumerate(augmented_images):
                                aug_output_path = os.path.join(output_person_dir, f"{folder_name}_{img_id}_aug_{idx+1}.jpg")
                                cv2.imwrite(aug_output_path, aug_image)

                        img_id += 1

            with open(track_file_path, "w") as track_file:
                track_file.write("This folder has been processed.")

            print(f"Processing complete for folder: {folder_name}!")

# Specify the base directory containing all the folders
base_directory = "Images"
output_directory = "Processed_Images"
new_width = 400  # Set the new width here

# Start processing all folders
process_folder(base_directory, output_directory, new_width)
```

# Dividing the dataset for training and testing using CNN for training And Testing

```python
import os
import shutil
from sklearn.model_selection import train_test_split

def split_dataset(base_dir, train_dir, test_dir, test_size=0.3):
    # Ensure train and test directories exist
    if not os.path.exists(train_dir):
        os.makedirs(train_dir)
    if not os.path.exists(test_dir):
        os.makedirs(test_dir)
    
    for folder_name in os.listdir(base_dir):
        folder_path = os.path.join(base_dir, folder_name)
        
        if os.path.isdir(folder_path):
            images = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

            if len(images) > 1:
                train_images, test_images = train_test_split(images, test_size=test_size, random_state=42)
            else:
                train_images = test_images = images  # If there's only one image, use it for both

            # Create train and test directories for this class
            class_train_dir = os.path.join(train_dir, folder_name)
            class_test_dir = os.path.join(test_dir, folder_name)
            os.makedirs(class_train_dir, exist_ok=True)
            os.makedirs(class_test_dir, exist_ok=True)
            
            for image in train_images:
                shutil.copy(os.path.join(folder_path, image), os.path.join(class_train_dir, image))
            
            for image in test_images:
                shutil.copy(os.path.join(folder_path, image), os.path.join(class_test_dir, image))

            print(f"Processed {folder_name} with {len(train_images)} train images and {len(test_images)} test images.")

# Example usage
base_directory = "Processed_Images"
train_directory = "Train_Images"
test_directory = "Test_Images"

split_dataset(base_directory, train_directory, test_directory)
```

# Remove Empty Folders

```python
# Function to remove empty folders
def remove_empty_folders(dataset_path):
    class_folders = os.listdir(dataset_path)
    for folder in class_folders:
        folder_path = os.path.join(dataset_path, folder)
        if os.path.isdir(folder_path):
            images = [f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff', '.webp'))]
            if not images:
                print(f"Removing empty folder: {folder_path}")
                os.rmdir(folder_path)

# Clean up any empty folders in Train_Images and Test_Images
remove_empty_folders('Train_Images')
remove_empty_folders('Test_Images')
```
