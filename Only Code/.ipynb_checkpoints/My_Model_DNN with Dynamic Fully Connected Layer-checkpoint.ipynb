# Modify the Model to Handle 128-Dimensional Embeddings

```python
class EmbeddingClassifierCNN(nn.Module):
    def __init__(self, num_classes):
        super(EmbeddingClassifierCNN, self).__init__()
        
        self.fc1 = nn.Linear(128, 256)
        self.dropout1 = nn.Dropout(0.3)  # Add dropout to avoid overfitting
        self.fc2 = nn.Linear(256, 128)
        self.dropout2 = nn.Dropout(0.3)
        self.fc3 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout1(x)
        x = F.relu(self.fc2(x))
        x = self.dropout2(x)
        x = self.fc3(x)  # Output layer
        return x
```

# Train the CNN Model Using Embeddings

```python
import numpy as np
import torch
import torch.optim as optim
import torch.nn.functional as F
import matplotlib.pyplot as plt
from torch.utils.data import TensorDataset, DataLoader

# Load embeddings and labels
encodings = np.load("E:\My thesis work\My trained Model\Simple_CNN_Resnet_face_encodings.npy")
labels = np.load("E:\My thesis work\My trained Model\Simple_CNN_Resnet_face_labels.npy")

# Convert to tensors
encodings_tensor = torch.tensor(encodings, dtype=torch.float32)
labels_tensor = torch.tensor(labels, dtype=torch.long)

# Check max label value and number of unique classes
print("Max label value:", labels.max())  # Ensure the max label is valid for classification
print("Number of unique labels:", len(set(labels)))  # Should match num_classes

# Ensure the number of classes is correctly set
num_classes = labels.max() + 1  # Ensure num_classes is one more than the max label value
print("Number of classes:", num_classes)

# Create a DataLoader
dataset = TensorDataset(encodings_tensor, labels_tensor)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Initialize model, criterion, and optimizer
model = EmbeddingClassifierCNN(num_classes=num_classes)
criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Lists to store loss and accuracy for plotting
train_losses = []
train_accuracies = []

# Training loop
num_epochs = 10
model.train()
for epoch in range(num_epochs):
    running_loss = 0.0
    correct_predictions = 0
    total_predictions = 0
    
    for i, (inputs, targets) in enumerate(dataloader):
        optimizer.zero_grad()
        outputs = model(inputs)
        
        # Calculate loss
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        
        # Calculate accuracy
        _, predicted = torch.max(outputs, 1)  # Get the index of the max log-probability
        correct_predictions += (predicted == targets).sum().item()
        total_predictions += targets.size(0)
    
    # Calculate the epoch accuracy and loss
    accuracy = 100 * correct_predictions / total_predictions
    train_losses.append(running_loss / len(dataloader))
    train_accuracies.append(accuracy)
    
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss / len(dataloader):.4f}, Accuracy: {accuracy:.2f}%")

# Plotting the training loss and accuracy
epochs = range(1, num_epochs + 1)

# Plot the training loss
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs')
plt.grid(True)

# Plot the training accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, train_accuracies, label='Training Accuracy', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Training Accuracy Over Epochs')
plt.grid(True)

plt.tight_layout()
plt.show()

# Save the trained model
torch.save(model.state_dict(), "embedding_classifier_cnn.pth")
```

```python
import numpy as np
import torch
import torch.optim as optim
import torch.nn.functional as F
import matplotlib.pyplot as plt
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import precision_score, recall_score, f1_score
import time  # Import time module for measuring processing time

# Load embeddings and labels
encodings = np.load("F:/Thesis latest work(6.12.24)/My thesis work/Face Encoding/Simple_CNN_Resnet_face_encodings.npy")
labels = np.load("F:/Thesis latest work(6.12.24)/My thesis work/Face Encoding/Simple_CNN_Resnet_face_labels.npy")

# Convert to tensors
encodings_tensor = torch.tensor(encodings, dtype=torch.float32)
labels_tensor = torch.tensor(labels, dtype=torch.long)

# Check max label value and number of unique classes
print("Max label value:", labels.max())  # Ensure the max label is valid for classification
print("Number of unique labels:", len(set(labels)))  # Should match num_classes

# Ensure the number of classes is correctly set
num_classes = labels.max() + 1  # Ensure num_classes is one more than the max label value
print("Number of classes:", num_classes)

# Create a DataLoader
dataset = TensorDataset(encodings_tensor, labels_tensor)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Initialize model, criterion, and optimizer
model = EmbeddingClassifierCNN(num_classes=num_classes)
criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Lists to store loss, accuracy, precision, recall, and f1 score for plotting
train_losses = []
train_accuracies = []
train_precisions_macro = []
train_precisions_micro = []
train_precisions_weighted = []
train_recalls = []
train_f1_scores = []

# Start measuring the total training time
overall_start_time = time.time()

# Training loop
num_epochs = 30
model.train()
for epoch in range(num_epochs):
    epoch_start_time = time.time()  # Start measuring time for this epoch

    running_loss = 0.0
    correct_predictions = 0
    total_predictions = 0
    true_labels = []
    predicted_labels = []
    
    epoch_process_start_time = time.time()  # Track total processing time for this epoch

    for i, (inputs, targets) in enumerate(dataloader):
        batch_start_time = time.time()  # Start measuring time for this batch

        optimizer.zero_grad()
        outputs = model(inputs)
        
        # Calculate loss
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

        # Calculate accuracy
        _, predicted = torch.max(outputs, 1)  # Get the index of the max log-probability
        correct_predictions += (predicted == targets).sum().item()
        total_predictions += targets.size(0)
        
        # Store true and predicted labels for precision, recall, and f1 score calculation
        true_labels.extend(targets.cpu().numpy())
        predicted_labels.extend(predicted.cpu().numpy())
        
        # Calculate processing time for this batch
        batch_end_time = time.time()
        batch_process_time = batch_end_time - batch_start_time
        if i == 0:  # Print process time for the first batch to check
            print(f"Processing time per batch: {batch_process_time:.4f} seconds")

    # Calculate the epoch accuracy, precision, recall, and f1 score
    accuracy = 100 * correct_predictions / total_predictions
    precision_macro = precision_score(true_labels, predicted_labels, average='macro', zero_division=1)
    precision_micro = precision_score(true_labels, predicted_labels, average='micro', zero_division=1)
    precision_weighted = precision_score(true_labels, predicted_labels, average='weighted', zero_division=1)
    recall = recall_score(true_labels, predicted_labels, average='macro', zero_division=1)
    f1 = f1_score(true_labels, predicted_labels, average='macro', zero_division=1)

    train_losses.append(running_loss / len(dataloader))
    train_accuracies.append(accuracy)
    train_precisions_macro.append(precision_macro)
    train_precisions_micro.append(precision_micro)
    train_precisions_weighted.append(precision_weighted)
    train_recalls.append(recall)
    train_f1_scores.append(f1)
    
    epoch_end_time = time.time()  # End measuring time for this epoch
    epoch_time = epoch_end_time - epoch_start_time  # Calculate time taken for this epoch
    
    epoch_process_end_time = time.time()  # Track total processing time for this epoch
    epoch_process_time = epoch_process_end_time - epoch_process_start_time  # Total process time for epoch
    
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss / len(dataloader):.4f}, "
          f"Accuracy: {accuracy:.2f}%, Macro Precision: {precision_macro:.4f}, Micro Precision: {precision_micro:.4f}, "
          f"Weighted Precision: {precision_weighted:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}, "
          f"Epoch Time: {epoch_time:.2f} seconds, Total Processing Time for Epoch: {epoch_process_time:.2f} seconds")

# End measuring the total training time
overall_end_time = time.time()
overall_training_time = overall_end_time - overall_start_time
print(f"Total training time: {overall_training_time:.2f} seconds")

# Plotting the training loss, accuracy, precision, recall, and f1 score
epochs = range(1, num_epochs + 1)

plt.figure(figsize=(15, 10))

# Plot the training loss
plt.subplot(2, 3, 1)
plt.plot(epochs, train_losses, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs')
plt.grid(True)

# Plot the training accuracy
plt.subplot(2, 3, 2)
plt.plot(epochs, train_accuracies, label='Training Accuracy', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Training Accuracy Over Epochs')
plt.grid(True)

# Plot the training precision
plt.subplot(2, 3, 3)
plt.plot(epochs, train_precisions_macro, label='Macro Precision', color='green')
plt.plot(epochs, train_precisions_micro, label='Micro Precision', color='red')
plt.plot(epochs, train_precisions_weighted, label='Weighted Precision', color='blue')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.title('Training Precision Over Epochs')
plt.grid(True)

# Plot the training recall
plt.subplot(2, 3, 4)
plt.plot(epochs, train_recalls, label='Training Recall', color='blue')
plt.xlabel('Epochs')
plt.ylabel('Recall')
plt.title('Training Recall Over Epochs')
plt.grid(True)

# Plot the training F1 score
plt.subplot(2, 3, 5)
plt.plot(epochs, train_f1_scores, label='Training F1 Score', color='red')
plt.xlabel('Epochs')
plt.ylabel('F1 Score')
plt.title('Training F1 Score Over Epochs')
plt.grid(True)

plt.tight_layout()
plt.show()

# Save the trained model
torch.save(model.state_dict(), "embedding_classifier_cnn.pth")

# Final comparison and display the best precision
final_macro_precision = train_precisions_macro[-1]
final_micro_precision = train_precisions_micro[-1]
final_weighted_precision = train_precisions_weighted[-1]

print(f"Final Macro Precision: {final_macro_precision:.4f}")
print(f"Final Micro Precision: {final_micro_precision:.4f}")
print(f"Final Weighted Precision: {final_weighted_precision:.4f}")

# Determine and display the best precision metric
best_precision_metric = "Macro Precision"
best_precision_value = final_macro_precision

if final_micro_precision > best_precision_value:
    best_precision_metric = "Micro Precision"
    best_precision_value = final_micro_precision

if final_weighted_precision > best_precision_value:
    best_precision_metric = "Weighted Precision"
    best_precision_value = final_weighted_precision

print(f"The best precision metric is {best_precision_metric} with a value of {best_precision_value:.4f}")
```

# Using the Pretrained CNN Model for Face Recognition

```python
import dlib
import cv2
import numpy as np
import joblib
import json
import torch

# Load the face detector and models
detector = dlib.get_frontal_face_detector()
sp = dlib.shape_predictor(r"E:\My thesis work\My trained Model\shape_predictor_68_face_landmarks.dat")
facerec = dlib.face_recognition_model_v1(r"E:\My thesis work\My trained Model\dlib_face_recognition_resnet_model_v1.dat")
classifier = joblib.load("face_recognition_model.pkl")  # SVC model

# Load the label-to-name dictionary
with open("label_to_name.json", "r") as f:
    label_to_name = json.load(f)

def recognize_face_with_svc(frame, detector, sp, facerec, model, label_to_name):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    for face in faces:
        # Get the landmarks and compute the face descriptor (embedding)
        landmarks = sp(gray, face)
        embedding = np.array(facerec.compute_face_descriptor(frame, landmarks))
        
        # Use the SVC model for prediction
        predicted_label = model.predict([embedding])[0]  # Predict the label

        # Get the person's name from the label
        name = label_to_name.get(str(predicted_label), "Unknown")

        # Display the name on the frame
        cv2.putText(frame, f"Name: {name}", (face.left(), face.top() - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        
        # Draw the bounding box around the face
        cv2.rectangle(frame, (face.left(), face.top()), (face.right(), face.bottom()), (0, 255, 0), 2)

    return frame

# Initialize video capture
cap = cv2.VideoCapture(1)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Recognize face using the SVC model
    frame = recognize_face_with_svc(frame, detector, sp, facerec, classifier, label_to_name)
    
    # Show the result
    cv2.imshow("Face Recognition", frame)

    # Exit on 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# My Custom CNN Confidence based Face Recognition

```python
import torch
import cv2
import dlib
import numpy as np
import json

# Load the trained CNN model
model = EmbeddingClassifierCNN(num_classes=num_classes)
model.load_state_dict(torch.load("embedding_classifier_cnn.pth"))
model.eval()  # Set to evaluation mode

# Load the face detector and dlib's facial landmarks and recognition models
detector = dlib.get_frontal_face_detector()
sp = dlib.shape_predictor(r"F:\Thesis latest work(6.12.24)\My thesis work\My trained Model\shape_predictor_68_face_landmarks.dat")
facerec = dlib.face_recognition_model_v1(r"F:\Thesis latest work(6.12.24)\My thesis work\My trained Model\dlib_face_recognition_resnet_model_v1.dat")

# Threshold for recognizing a face as "unknown"
CONFIDENCE_THRESHOLD = 0.2

def recognize_face_with_cnn(frame, detector, sp, facerec, model, label_to_name):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    for face in faces:
        # Get the landmarks and compute the face descriptor (embedding)
        landmarks = sp(gray, face)
        embedding = np.array(facerec.compute_face_descriptor(frame, landmarks))

        # Convert the embedding to a tensor and reshape it
        embedding_tensor = torch.tensor(embedding, dtype=torch.float32).unsqueeze(0)  # Add batch dimension
        
        # Use the CNN model for prediction
        with torch.no_grad():
            output = model(embedding_tensor)  # Get the raw logits
            probabilities = torch.softmax(output, dim=1)  # Convert logits to probabilities
            confidence, predicted_label = torch.max(probabilities, 1)  # Get confidence and label
        
        # Determine if the confidence is above the threshold
        if confidence.item() < CONFIDENCE_THRESHOLD:
            name = "Unknown"
        else:
            # Get the person's name from the label
            name = label_to_name.get(str(predicted_label.item()), "Unknown")

        # Display the name on the frame
        cv2.putText(frame, f"Name: {name} ({confidence.item():.2f})", (face.left(), face.top() - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        # Draw the bounding box around the face
        cv2.rectangle(frame, (face.left(), face.top()), (face.right(), face.bottom()), (0, 255, 0), 2)

    return frame

# Initialize video capture
cap = cv2.VideoCapture(0)

# Load label-to-name dictionary (map of label to person name)
with open("label_to_name.json", "r") as f:
    label_to_name = json.load(f)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Recognize face using the CNN model
    frame = recognize_face_with_cnn(frame, detector, sp, facerec, model, label_to_name)
    
    # Show the result
    cv2.imshow("Face Recognition", frame)

    # Exit on 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Pretrained Confidence Based Recognition

```python
import dlib
import cv2
import numpy as np
import joblib
import json

# Load the face detector and models
detector = dlib.get_frontal_face_detector()
sp = dlib.shape_predictor(r"F:\Thesis latest work(6.12.24)\My thesis work\PRETRAINED DLIB Resnet\shape_predictor_68_face_landmarks.dat")
facerec = dlib.face_recognition_model_v1(r"F:\Thesis latest work(6.12.24)\My thesis work\PRETRAINED DLIB Resnet\dlib_face_recognition_resnet_model_v1.dat")
classifier = joblib.load("face_recognition_model.pkl")  # SVC model

# Load the label-to-name dictionary
with open("label_to_name.json", "r") as f:
    label_to_name = json.load(f)

# Threshold for recognizing "unknown" faces
CONFIDENCE_THRESHOLD = 0.6


def compute_softmax(logits):
    """Compute softmax probabilities from raw logits."""
    exp_logits = np.exp(logits - np.max(logits))
    return exp_logits / exp_logits.sum()


def recognize_face_with_svc(frame, detector, sp, facerec, model, label_to_name, threshold):
    """
    Recognize faces using the SVC model.
    """
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    for face in faces:
        # Get landmarks and face descriptor (embedding)
        landmarks = sp(gray, face)
        embedding = np.array(facerec.compute_face_descriptor(frame, landmarks))
        
        # Predict probabilities using the SVC model
        logits = model.decision_function([embedding])  # Raw scores
        probabilities = compute_softmax(logits[0])  # Softmax probabilities
        max_confidence = np.max(probabilities)  # Highest confidence score
        predicted_label = np.argmax(probabilities)  # Label with max confidence
        
        # Determine the name based on confidence threshold
        if max_confidence < threshold:
            name = "Unknown"
        else:
            name = label_to_name.get(str(predicted_label), "Unknown")
        
        # Display the name and confidence on the frame
        cv2.putText(frame, f"Name: {name} ({max_confidence:.2f})", 
                    (face.left(), face.top() - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        
        # Draw a bounding box around the face
        cv2.rectangle(frame, (face.left(), face.top()), 
                      (face.right(), face.bottom()), (0, 255, 0), 2)

    return frame


# Initialize video capture
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Recognize face using the updated SVC model
    frame = recognize_face_with_svc(frame, detector, sp, facerec, classifier, label_to_name, CONFIDENCE_THRESHOLD)
    
    # Show the result
    cv2.imshow("Face Recognition", frame)

    # Exit on 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
